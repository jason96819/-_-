## 정규 분포

```python
from scipy.stats import norm
```

```python
# 평균과 표준편차
mu = 0
sigma = 1

# 확률밀도함수 계산
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
pdf = norm.pdf(x, mu, sigma)

# 그래프 그리기
plt.plot(x, pdf)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.grid(True)
plt.show()
```

```python
# 평균과 표준편차
mu = [0, 1, 2, 3, 4, 5]
sigma = 1

for i in mu:
    # 확률밀도함수 계산
    x = np.linspace(i - 3 * sigma, i + 3 * sigma, 100)
    pdf = norm.pdf(x, i, sigma)

    # 그래프 그리기
    plt.plot(x, pdf, label=f'mu = {i}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 평균과 표준편차
mu = 0
sigma = [1, 2, 3, 4, 5]

for i in sigma:
    # 확률밀도함수 계산
    x = np.linspace(mu - 3 * i, mu + 3 * i, 100)
    pdf = norm.pdf(x, mu, i)

    # 그래프 그리기
    plt.plot(x, pdf, label=f'sigma = {i}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.grid(True)
plt.legend()
plt.show()
```

## 지수 분포

```python
from scipy.stats import expon
```

```python
# 파라미터 lambda
lmbda = [0.1, 0.3, 0.5, 0.7, 0.9]

for i in lmbda:
    # 확률밀도함수 계산
    x = np.linspace(0, 10, 100)
    pdf = expon.pdf(x, scale=1/i)

    # 그래프 그리기
    plt.plot(x, pdf, label=f'lambda = {i}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution')
plt.grid()
plt.legend()
plt.show()
```
